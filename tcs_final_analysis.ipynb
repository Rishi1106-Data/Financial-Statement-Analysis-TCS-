# 📊 TCS Financial Analysis Project

Complete analysis using final dataset.

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("final_dataset.csv")

print(df.columns)
df.head()
```

```
df.columns = df.columns.str.strip() \
                       .str.lower() \
                       .str.replace(" ", "_") \
                       .str.replace("+", "") \
                       .str.replace("%", "percent")

print(df.columns)
```

```
def get_col(df, keywords):
    for col in df.columns:
        for key in keywords:
            if key in col:
                return col
    return None

sales_col = get_col(df, ['sales', 'revenue'])
profit_col = get_col(df, ['profit'])
expense_col = get_col(df, ['expense'])
cash_col = get_col(df, ['cash', 'operating'])

print(sales_col, profit_col, expense_col, cash_col)
```

```
df['profit_margin'] = df[profit_col] / df[sales_col]
df['expense_ratio'] = df[expense_col] / df[sales_col]
```

```
plt.figure()
plt.plot(df['quarter'], df[sales_col], marker='o')
plt.title("Revenue Trend")
plt.xticks(rotation=45)
plt.savefig("revenue_trend.png")
plt.show()
```

```
plt.figure()
plt.plot(df['quarter'], df[profit_col], marker='o')
plt.title("Profit Trend")
plt.xticks(rotation=45)
plt.savefig("profit_trend.png")
plt.show()
```

```
fig, ax1 = plt.subplots()

ax1.plot(df['quarter'], df[profit_col], marker='o')
ax1.set_ylabel("Profit")

if cash_col:
    ax2 = ax1.twinx()
    ax2.plot(df['quarter'], df[cash_col], marker='o')
    ax2.set_ylabel("Cash Flow")

plt.title("Profit vs Cash Flow")
plt.xticks(rotation=45)
plt.savefig("profit_vs_cashflow.png")
plt.show()
```

```
plt.figure()
plt.plot(df['quarter'], df['profit_margin'], marker='o')
plt.title("Profit Margin")
plt.xticks(rotation=45)
plt.savefig("profit_margin.png")
plt.show()
```

```
df.to_csv("final_dataset_cleaned.csv", index=False)
```

## 📌 Key Insights

- Revenue shows consistent growth indicating strong business expansion  
- Profit trend remains stable with minor fluctuations  
- Cash flow differs from profit, indicating working capital effects  
- Expense ratio is controlled, showing operational efficiency  
- Overall financial position is strong and stable
